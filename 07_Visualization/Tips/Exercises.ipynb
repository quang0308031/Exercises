### Step 1. Import the necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv')
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
del(tips['Unnamed: 0'])
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(tips['total_bill'], bins = 25)
plt.xlabel('total_bill')
plt.ylabel('frequency')
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.scatter(tips.total_bill, tips.tip)
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.scatter(tips.total_bill, tips.tip, s=tips.size / 10, alpha=0.4)
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.xlabel('total_bill')
plt.ylabel('day')
plt.scatter(tips.total_bill, tips.day)
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

tips.sex = label.fit_transform(tips.sex)
```

```python
plt.xlabel('tip')
plt.ylabel('day')
dot = plt.scatter(tips[tips.sex == 0].tip, tips[tips.sex == 0].day, alpha = 0.7, c='red')
dot1 = plt.scatter(tips[tips.sex == 1].tip, tips[tips.sex == 1].day, alpha = 0.7, c='blue')
plt.legend(["FeMale", "Male"])
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
print(tips.time.unique())
```

```python
data1 = tips.total_bill
data2 = tips[tips.time == 'Dinner'].total_bill
data3 = tips[tips.time == 'Lunch'].total_bill
data = [data1, data2, data3]
x_ticks = ['Day', 'Dinner', 'Lunch']
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("Time")
ax.set_ylabel("total_bill")
ax.set_xticklabels(x_ticks)
ax.boxplot(data)
plt.show()

```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
df1 = tips[tips.time == 'Dinner'].tip
df2 = tips[tips.time == 'Lunch'].tip
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize=(30, 10))
ax1.set_title('Dinner')
ax1.set_xlabel('total_bill')
ax1.set_ylabel('frequency')
ax1.hist(df1)

ax2.set_title('Lunch')
ax2.set_xlabel('total_bill')
ax2.set_ylabel('frequency')
ax2.hist(df2)

plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
tips.smoker = label.fit_transform(tips.smoker)
tips.head()
```

```python
temp1 = tips[tips.sex == 0] #Female
temp2 = tips[tips.sex == 1] #Male
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize=(30, 10))
ax1.set_title('Female')
ax1.set_xlabel('total_bill')
ax1.set_ylabel('tip')
dot1 = ax1.scatter(temp1[temp1.smoker == 0].total_bill, temp1[temp1.smoker == 0].tip, c='green')
dot2 = ax1.scatter(temp1[temp1.smoker == 1].total_bill, temp1[temp1.smoker == 1].tip, c='blue')
ax1.legend(['non-smoker', 'smoker'])
ax1.grid()

ax2.set_title('Male')
ax2.set_xlabel('total_bill')
ax2.set_ylabel('tip')
dot1 = ax2.scatter(temp2[temp2.smoker == 0].total_bill, temp2[temp2.smoker == 0].tip, c='green')
dot2 = ax2.scatter(temp2[temp2.smoker == 1].total_bill, temp2[temp2.smoker == 1].tip, c='blue')
ax2.legend(['non-smoker', 'smoker'])
ax2.grid()

plt.show()
```
